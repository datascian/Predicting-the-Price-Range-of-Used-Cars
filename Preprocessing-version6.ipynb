#### Preprocessing 1 (P1) = Remove the following columns: 
'url' - does not affect the price

'region_url' - because web address of the region where it being sold does not affect the price

'VIN' - the vehicle identification number is a unique code that is used to identify where and when the vehicle was made. This does not affect price.

'lat' and 'long' - because location is already indicated by the region

'image_url' and 'description' - It was planned to transform this into a binary variable indicating whether this information was provided. The model would have determined whether having these information would affect the price or not. However, approximately all the rows have this information making these binary variables inconsequential to the model


```python
raw_data_p1 = raw_data.drop(['url','region_url','VIN','lat','long','image_url', 'description'],axis=1)
```

```python
raw_data_p1.info(null_counts=True)
```

#### P2 = Move price to the first column

```python
raw_data_p2 = raw_data_p1.copy()
price_column = raw_data_p2.pop('price')
raw_data_p2.insert(0, 'price', price_column)
raw_data_p2.info(null_counts=True)
```

#### P3 = Remove row with missing data in any of the independent variable

```python
#Get the list of independent variables
independent_variables = raw_data_p2.iloc[:,1:-1].columns.tolist()
independent_variables
```

```python
raw_data_p3 =raw_data_p2.dropna(how='any', subset=independent_variables)
raw_data_p3.shape
```

```python
#get the number of missing values per column and the percentage to the total number of records
print("column      number of missing values      percentage")
column_names = list(raw_data_p3.columns)
for c in column_names:
    num_missing_values = sum(raw_data_p3[c].isnull())
    percentage = round(num_missing_values/raw_data_p3.shape[0]*100, 2)
    print ("%-15s   %10s           %10.2f" % (c, f"{num_missing_values:,}", percentage))
```

#### P4 = Clean the 'model' column by changing all values to lower caps and removing white spaces

```python
#Check the number of unique values for 'model'
raw_data_p4 = raw_data_p3.copy()
raw_data_p3['model'].nunique()
```

```python
#make everything in the model lowercase
raw_data_p4['model'] = raw_data_p4['model'].str.lower()
```

```python
#remove white spaces from column 'model'
raw_data_p4['model'] = raw_data_p4['model'].str.strip()
```

#### P5 = Remove the outliers in price and odometer by using the IQR rule

```python
#Determine the first quartile and third quartile and compute the inter-quartile range 
Q1 = raw_data_p4.quantile(0.25)
Q3 = raw_data_p4.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(Q1)
print(Q3)
```

```python
#Remove the outliers from the dataset and check how may rows are left
raw_data_p5 = raw_data_p4[~((raw_data_p4 < (Q1 - 1.5 * IQR))|(raw_data_p4 > (Q3 + 10 * IQR))).any(axis=1)]
raw_data_p5.shape
```

```python
#Number of rows removed in P5
print("Number of rows removed in P5: ", raw_data_p4.shape[0]-raw_data_p5.shape[0])
#percentage of rows of original dataset removed
print("Percentage: ", round(((raw_data_p4.shape[0]-raw_data_p5.shape[0])/raw_data.shape[0]*100),2))
raw_data_p5.info()
```

```python
raw_data_p5.describe()
```

#### P11 = Convert odometer into a categorical variable

```python
#check the minimum and maximum values
raw_data_p10.describe()
```

```python
#create a list of bins for the ranges
odo_bins = []
for i in range(32):
    odo_bins.append(12000*i)
odo_bins
```

```python
#create a list of labels for the ranges
odo_labels = ['0-12k', '12k-24k', '24k-36k', '36k-48k', '48k-60k', '60k-72k', '72k-84k', '84k-96k', '96k-108k', '108k-120k',
            '120k-132k', '132k-144k', '144k-156k', '156k-168k', '168k-180k', '180k-192k', '192k-204k', '204k-216k',
            '216k-228k', '228k-240k', '240k-252k', '252k-264k', '264k-276k', '276k-288k', '288k-300k', '300k-312k',
            '312k-324k', '324k-336k', '336k-348k', '348k-360k', '360k-372k']
```

```python
#Create the new column 'odometer_range'
raw_data_p11 = raw_data_p10.copy()
odo_category = pd.cut(raw_data_p11.odometer,bins=odo_bins,labels=odo_labels)
raw_data_p11.insert(9,'odometer_range',odo_category)
raw_data_p11.info()
```

```python
"""The above table shows that 'odometer_range' has some null values. We check the rows which
have null values in this column."""
na_odo_range = raw_data_p11[raw_data_p11['odometer_range'].isna()]
na_odo_range
```

```python
"""The rows will null values for 'odometer_range' are the rows which an odometer reading of 0. 
We categorize these rows as '0-12k' as well."""
raw_data_p11['odometer_range'] = raw_data_p11['odometer_range'].fillna('0-12k')
raw_data_p11.info()
```

```python
raw_data_p11['odometer_range'].nunique()
```

```python
#Remove the column 'odometer'
raw_data_p11 = raw_data_p11.drop(['odometer'], axis=1)
raw_data_p11.info()
```

#### P12 = Change the 'price' column to a categorical variable

```python
#check the minimum and maximum values
raw_data_p11.describe()
```

```python
#create a list of bins for the ranges
price_bins = []
for i in range(28):
    price_bins.append(5000*i)
price_bins
```

```python
#create a list of labels for the ranges
price_labels = ['0-5k', '5k-10k', '10k-15k', '15k-20k', '20k-25k', '25k-30k', '30k-35k', '35k-40k', '40k-45k', '45k-50k',
               '50k-55k', '55k-60k', '60k-65k', '65k-70k', '70k-75k', '75k-80k', '80k-85k', '85k-90k', '90k-95k', '95k-100k',
               '100k-105k', '105k-110k', '110k-115k', '115k-120k', '120k-125k', '125k-130k', '130k-135k']
```

```python
len(price_labels)
```

```python
#Create the new column 'price_range'
raw_data_p12 = raw_data_p11.copy()
price_category = pd.cut(raw_data_p12.price,bins=price_bins,labels=price_labels)
raw_data_p12.insert(1,'price_range',price_category)
raw_data_p12
```

```python
#Remove the column 'price'
raw_data_p12 = raw_data_p12.drop(['price'], axis=1)
raw_data_p12.info()
```

```python
#Export this to a csv file for exploratory data analysis
raw_data_p12.to_csv('preprocessed_data_eda.csv', index=False)
```

#### P13 = One-hot encode the categorical variables

```python
independent_variables = raw_data_p12.iloc[:,1:-1].columns.tolist()
independent_variables
```

```python
for c in independent_variables:
    print (c, raw_data_p12[c].nunique())
```

```python
#one-hot encode the categorical variables
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore')
encoded_data = pd.DataFrame(ohe.fit_transform(raw_data_p12[independent_variables]).toarray())
encoded_data.columns = ohe.get_feature_names(independent_variables)
encoded_data
```

```python
#remove the original columns for the categorical variables in the dataframe
raw_data_p13 = raw_data_p12.drop(independent_variables,axis=1)
raw_data_p13
```

```python
#add the encoded categorical variables to the dataframe
encoded_data.reset_index(drop=True, inplace=True)
raw_data_p13.reset_index(drop=True, inplace=True)
combined_data = pd.concat([raw_data_p13, encoded_data],axis=1)
combined_data
```
