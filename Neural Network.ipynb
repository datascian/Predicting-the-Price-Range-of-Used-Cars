#### Import the relevant libraries

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras import metrics
from time import time
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

#### Load the data

```python
#train_data = np.loadtxt('train_data.csv', delimiter=',')
train_data = pd.read_csv('train_data.csv', index_col=False)
train_data
```

```python
train_data.shape
```

#### Separate the dependent variable from the independent variables in the whole data

```python
X_train = train_data.iloc[:,1:].values
y_train = train_data.iloc[:, 0].values
```

#### Train the model with k-fold cross validation

```python
#check the start time in order to calculate the runtime of training the model
start = time()

#Define 5-fold cross validation test
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
cvscores=[]
for train, test in cv.split(X_train, y_train):
    #Create model
    hidden_layer_size = 1500
    output_size= 27
    model = Sequential()
    model.add(Dense(hidden_layer_size, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(hidden_layer_size, activation='relu'))
    model.add(Dense(output_size, activation='softmax'))
    #Compile the model
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    #Fit the model
    model.fit(X_train, y_train, epochs=50, batch_size=1500)
    #Evaluate the model
    scores = model.evaluate(X_train[test], y_train[test], verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)

#check the end time
end = time()
print("The runtime of the model training:", end-start)

print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```

#### Fit the model

```python
# define the keras model
hidden_layer_size = 1500
output_size= 27
model = Sequential()
model.add(Dense(hidden_layer_size, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(hidden_layer_size, activation='relu'))
model.add(Dense(output_size, activation='softmax'))

# compile the keras model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#check the start time in order to calculate the runtime of training the model
start = time()

# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=50, batch_size=1500)

#check the end time
end = time()
print("The runtime of the model training:", end-start)
```

#### Load the test data

```python
test_data = pd.read_csv('test_data.csv')
test_data
```

#### Separate the dependent variable from the independent variables in the test data

```python
X_test = test_data.iloc[:,1:].values
y_test = test_data.iloc[:, 0].values
```

#### Make predictions on the test data using the model

```python
test_prediction = np.argmax(model.predict(X_test),axis=1)
```

#### Evaluate the performance of the model by checking the precision, recall and overall accuracy

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100)) 
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, test_prediction, digits=4)
print(report)
```
